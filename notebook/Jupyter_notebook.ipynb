## Data Preprocessing

```python
# changing float to integer
data["bathrooms"] = data["bathrooms"].astype(int)
data["floors"] = data["floors"].astype(int)
```

```python
# renaming the column yr_built to age and changing the values to age
data.rename(columns={'yr_built':'age'},inplace = True)
```

```python
from datetime import datetime
```

```python
today = datetime.now()
```

```python
data["age"] = today.year - data["age"]
```

```python
data["age"].head()
```

```python
# changing the column yr_renovated to renovated and changing the values to 0 and a
data.rename(columns={'yr_renovated':'renovated'},inplace=True)
data['renovated'] = data['renovated'].apply(lambda x: 0 if x == 0 else 1)
```

```python
data["renovated"].head()
```

```python
# using simple feature scaling
# List of columns to scale
columns_to_scale = ['sqft_living', 'sqft_living15', 'sqft_lot', 'sqft_above', 'sqft_basement', 'sqft_lot15']

# Apply scaling
data[columns_to_scale] = data[columns_to_scale].apply(lambda x: x / x.max())
```

#### Plotting the location of the houses based on longitude and latitude on the map

```python
# adding a new column price_range and categorizing the price into 4 categories
data["price_range"] = pd.cut(data["price"],bins = [0,321950,450000,645000,1295648],labels=['Low','Medium','High','Very High'])
```

```python
map = folium.Map(location=[47.5480, -121.9836],zoom_start=8)
marker_cluster = FastMarkerCluster(data[['lat', 'long']].values.tolist()).add_to(map)
map
```

```python
data = data.drop("price_range",axis=1)
```

### Preparing data

```python
x = data.drop("price",axis=1)
y = data["price"]
```

### Train test split

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state=10)
```

### Model Train

#### Using pipeline to combine the transformers and estimators and fit the model

```python
input = [('scale',StandardScaler()),('polynomial', PolynomialFeatures(degree=2)),('model',LinearRegression())]
pipe = Pipeline(input)
pipe
```

```python
pipe.fit(x_train,y_train)
```

```python
pipe.score(x_test,y_test)
```

```python
pipe_pred = pipe.predict(x_test)
```

```python
r2_score(pipe_pred,y_test)
```

### Ridge Regression

```python
Ridgemodel = Ridge(alpha = 0.001)
Ridgemodel
```

```python
# training the model
Ridgemodel.fit(x_train,y_train)
Ridgemodel.score(x_test,y_test)
```

```python
#testing the model
r_pred = Ridgemodel.predict(x_test)
r2_score(y_test,r_pred)
```

### Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=100, random_state=42)
regressor
```

```python
# training the model
regressor.fit(x_train,y_train)
regressor.score(x_test,y_test)
```

```python
#testing the model
yhat = regressor.predict(x_test)
r2_score(y_test,yhat)
```

## Model Evalution

### Distribution plot from the models predictions and the actual values

```python
# Updated plot with filled KDE plots for actual and predicted prices
fig, ax = plt.subplots(1, 3, figsize=(20, 5))

# Plotting KDE plots with fill
sns.kdeplot(y_test, ax=ax[0], color="blue",  label='Actual Price')
sns.kdeplot(pipe_pred, ax=ax[0], color="red",  label='Predicted Price')

sns.kdeplot(y_test, ax=ax[1], color="blue", label='Actual Price')
sns.kdeplot(r_pred, ax=ax[1], color="red",  label='Predicted Price')

sns.kdeplot(y_test, ax=ax[2], color="blue", label='Actual Price')
sns.kdeplot(yhat, ax=ax[2], color="red",  label='Predicted Price')

# Adding legends
ax[0].legend(['Actual Price', 'Predicted Price'])
ax[1].legend(['Actual Price', 'Predicted Price'])
ax[2].legend(['Actual Price', 'Predicted Price'])

# Model names as titles
ax[0].set_title('Linear Regression')
ax[1].set_title('Ridge Regression')
ax[2].set_title('Random Forest Regression')

plt.show()
```

### Error Evaluation

```python
#plot the graph to compare mae, mse, rmse for all models
fig, ax = plt.subplots(1,3,figsize=(20,5))
sns.barplot(x=['Linear Regression','Ridge Regression','Random Forest'],y=[mean_absolute_error(y_test,pipe_pred),mean_absolute_error(y_test,r_pred),mean_absolute_error(y_test,yhat)],ax=ax[0])
sns.barplot(x=['Linear Regression','Ridge Regression','Random Forest'],y=[mean_squared_error(y_test,pipe_pred),mean_squared_error(y_test,r_pred),mean_squared_error(y_test,yhat)],ax=ax[1])
sns.barplot(x=['Linear Regression','Ridge Regression','Random Forest'],y=[np.sqrt(mean_squared_error(y_test,pipe_pred)),np.sqrt(mean_squared_error(y_test,r_pred)),np.sqrt(mean_squared_error(y_test,yhat))],ax=ax[2])
# label for the graph
ax[0].set_ylabel('Mean Absolute Error')
ax[1].set_ylabel('Mean Squared Error')
ax[2].set_ylabel('Root Mean Squared Error')
plt.show()
```

### Accuracy Evaluation

```python
# plot accuracy of all models in the same graph
fig, ax = plt.subplots(figsize=(7,5))
sns.barplot(x=['Linear Regression','Ridge Regression','Random Forest Regression'],y=[metrics.r2_score(y_test,pipe_pred),metrics.r2_score(y_test,r_pred),metrics.r2_score(y_test,yhat)])
ax.set_title('Accuracy of all models')
plt.show()
```

```python
## House Price Prediction Case Study Summary

In this case study, I applied several regression algorithms to predict house prices based on a dataset containing various features. The performance of each algorithm was evaluated using training scores and R^2 scores, which measure how well the model explains the variance in the target variable (house prices) Below are the results:

| Algorithm                  | Training Score | R^2 Score (Testing) |
|----------------------------|----------------|---------------------|
| Linear Regression           | 0.82           | 0.79                |
| Ridge Regression            | 0.71           | 0.71                |
| Random Forest Regressor     | 0.87           | 0.87                |

### Insights:

- **Linear Regression**: The model performed well, achieving a training score of 0.82 and an R^2 score of 0.79 on the test data. This suggests that Linear Regression captures the relationship between the features and house prices effectively but leaves room for improvement.

- **Ridge Regression**: Ridge Regressaion, a regularized version of Linear Regression, achieved slightly lower scores (0.71 for both training and testing). This indicates that the model may generalize better than standard Linear Regression, but its predictive power is somewhat limited by the data or feature interactions.

- **Random Forest Regressor**: The Random Forest Regressor outperformed the other models, with a training score of 0.87 and an R^2 score of 0.87 on the test set. This suggests that the model handles complex feature interactions better, making it more effective in predicting house prices.

### Conclusion:

Overall, the Random Forest Regressor delivered the best performance, both in terms of training and testing accuracy, indicating that it captured the complex patterns in the data more effectively. While Linear Regression also performed reasonably well, Ridge Regression regularization made it more resistant to overfitting, though at the cost of lower accuracy. Future improvements might involve tuning hyperparameters or incorporating feature engineering to enhance the models performance.
```
